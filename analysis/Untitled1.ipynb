```python
import numpy as np
import pandas as pd
from time import time
import numpy_groupies as npg
```

```python
import pickle
d = None
with open('data/dbg.dmp','rb') as f:
    d = pickle.load(f)
```

```python
cc = ['frame','_error','hml_x','hml_y','hml_z','hmr_x','hmr_y','hmr_z']
def rf(idx):
    if (idx[0]%100)==0:
        print( " Processed "+str(idx[0])+"/"+str(d.shape[0])+" rows for anomalies" )
    df = d.ix[idx,cc]
    df['_error'] = np.sqrt(np.square(df[cc] - row[cc]).sum(axis=1))
    return df.groupby('frame').min().sum()['_error']
    
d['_error'] = np.nan
a = pd.rolling_apply(d, window=1000, func=rf, center=True)
```

```python
cc = ['frame','hml_x','hml_y','hml_z','hmr_x','hmr_y','hmr_z']
lastInfoTime = time()
for irow, row in d.iterrows():
    if (time() - lastInfoTime) > 2.0:
        lastInfoTime = time()
        print( " Processed "+str(irow)+"/"+str(d.shape[0])+" rows for anomalies" )
    rc = (d.frame>=row.frame-300) & (d.frame<row.frame+300)
    d.ix[] = np.sqrt(np.square(d[rc][cc] - row[cc]).sum(axis=1))
    row['hm_e_h'] = d[rc].groupby('frame').min().sum()['error']
```

```python
cc = ['frame','hml_x','hml_y','hml_z','hmr_x','hmr_y','hmr_z']
lastInfoTime = time()
dnpt = d[cc].as_matrix()
dnp  = np.full( (dnpt.shape[0], dnpt.shape[1]+1), np.nan)
dnp[:,:-1] = dnpt
wsl = 10
wsr = 0
for irow in range(wsl,dnp.shape[0]-wsr):
    if (time() - lastInfoTime) > 2.0:
        lastInfoTime = time()
        print( " Processed "+str(irow)+"/"+str(d.shape[0])+" rows for anomalies" )
    f = dnp[(irow-wsl):(irow+wsr),0].astype('int')
    e = np.linalg.norm(dnp[(irow-wsl):(irow+wsr),1:7] - dnp[irow,1:7], axis=1)
    dnp[irow,7] = sum(npg.aggregate(f,e, func='min'))
```

```python
cc = ['hml_x','hml_y','hml_z','hmr_x','hmr_y','hmr_z']
lastInfoTime = time()
d.frame = d.frame.astype('int')
d.set_index('frame')
for irow, row in d.iterrows():
    if (time() - lastInfoTime) > 2.0:
        lastInfoTime = time()
        print( " Processed "+str(irow)+"/"+str(d.shape[0])+" rows for anomalies" )
    rc = (d.frame>=row.frame-100) & (d.frame<row.frame+100)
    row['hm_e_h'] = sum(npg.aggregate(d.ix[rc,'frame'], np.sqrt(np.square(d[rc][cc] - row[cc]).sum(axis=1)), func='min'))
```
